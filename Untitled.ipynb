```python

import pandas as pd
pd.set_option('display.max_columns', 20)
pd.set_option('display.max_rows', 20)
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.gridspec import GridSpec
#np.random.seed(1)

```

```python
number_of_samples = 200
X = np.abs(np.random.uniform(0,6,(number_of_samples,2))) #create random X
X[:,0] = 1
X = pd.DataFrame(X,columns=['Intercept','wk'])
X['wk2'] = X['wk']**2
X= X.sort_values(by='wk').reset_index(drop=True)
print (X)
```

```python
def exponential(x):
    return 2.718**x

# assume that the model of plant growth is 
X['theoretical_weight'] =  exponential(3*X['Intercept']+0.53*X['wk']-0.06*X['wk2'])

# add random noise, ignore the specifics
scale = 0.04 #compare this to scale in model later
noise = scale**.5*np.random.normal(0,X['theoretical_weight'],X['theoretical_weight'].shape)

weight_noisy = X['theoretical_weight'] + noise
X['measured_weight'] = weight_noisy
print (X)
plt.xlabel('Weeks')
plt.ylabel('Weight of plant')
plt.plot(X['wk'],X['measured_weight'],'ro',label='Measured weight')
plt.plot(X['wk'],X['theoretical_weight'],'b',label='Theoretical weight')
plt.legend()
plt.show()
```

```python
resultsgamma = smf.glm('measured_weight ~ -1 +Intercept +  wk +wk2   ',
                  data=X,
                  family=sm.families.Gamma(link = sm.genmod.families.links.log)).fit();
print (resultsgamma.summary())
X['Prediction'] = resultsgamma.predict(X)
X['Variance_prediction'] = np.diag(np.matmul(np.matmul(X[['Intercept','wk','wk2']].values,resultsgamma.cov_params().values),X[['Intercept','wk','wk2']].values.T))
X['Std_prediction'] = (1-np.exp(X['Variance_prediction']**.5))
X['prediction_lower5_confidence_interval'] = np.exp(np.log(X['Prediction'])-2*X['Std_prediction'])
X['prediction_upper95_confidence_interval'] = np.exp(np.log(X['Prediction'])+2*X['Std_prediction'])


#plt.plot(X['wk'],X['measured_weight'],'ro',label='Measured weight')
plt.plot(X['wk'],X['theoretical_weight'],'b',label='Theoretical weight')
plt.plot(X['wk'],X['Prediction'],'g',label='Predicted weight and confidence interval 95')
plt.plot(X['wk'],X['prediction_lower5_confidence_interval'].values,color='g',ls='--',label=None)
plt.plot(X['wk'],X['prediction_upper95_confidence_interval'].values,color='g',ls='--',label=None)
plt.xlabel('Weeks')
plt.ylabel('Weight of plant')
plt.legend()
plt.show()
```
